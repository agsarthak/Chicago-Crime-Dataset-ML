```python
dfEmpty=CrimeData.isnull().sum()
print(dfEmpty)
```

```python
def percentEmpty(df):
    dfEmpty=df.isnull().sum()
    totalRows=len(df)
    print("totalRows",totalRows)
    peDf = pd.DataFrame((dfEmpty/totalRows)*100, columns=['percent'])
    peDf['colName'] = peDf.index
    return peDf
```

```python
percentEmpty(CrimeData)
```

```python
CrimeData.dropna(subset=['Case Number', 'District', 'Latitude'], how='any', inplace=True)
```

```python
percentEmpty(CrimeData)
```
